```python
#Finding the most Popular Movie

# The merged_df1_df2 has a column named 'title' containing movie titles

def most_popular_movie(df):
    # Count occurrences of each movie title
    movie_counts = df['title'].value_counts()
    # Find the most popular movie
    most_popular_movie = movie_counts.idxmax()
    return most_popular_movie

most_popular = most_popular_movie(merged_df1_df2)
print("The most popular movie is:", most_popular)
```

```python

def most_popular_movies(df, n=5):
    # Count occurrences of each movie title
    movie_counts = df['title'].value_counts()
    # Find the top n most popular movies
    top_n_movies = movie_counts.head(n).index.tolist()
    return top_n_movies

top_5_popular = most_popular_movies(merged_df1_df2, n=5)
print("The top 5 most popular movies are:", top_5_popular)
```

```python
#To get the least popular movies
def least_popular_movies(df, n=5):
    # Count occurrences of each movie title
    movie_counts = df['title'].value_counts()
    # Find the bottom n least popular movies
    least_n_movies = movie_counts.tail(n).index.tolist()
    return least_n_movies

least_5_popular = least_popular_movies(merged_df1_df2, n=5)
print("The bottom 5 least popular movies are:", least_5_popular)
```

```python
#To get the most popular movie in a chart format
def most_popular_movies(df, n=5):
    # Count occurrences of each movie title
    movie_counts = df['title'].value_counts()
    # Find the top n most popular movies
    top_n_movies = movie_counts.head(n)
    return top_n_movies

top_5_popular = most_popular_movies(merged_df1_df2, n=5)
# Create a bar chart
plt.bar(top_5_popular.index, top_5_popular.values)
plt.xlabel('Title')
plt.ylabel('Frequency')
plt.title('Top 5 Most Popular Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#After this analysis above, we can see that the highest rated movie is Eden and Truth and dare with a frequency of 7 and 5 respectively
```
